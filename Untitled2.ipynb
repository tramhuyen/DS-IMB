```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
link = 'https://drive.google.com/file/d/19ULyINxxSMlUQ9gssX6J4Be0OBuFgvB6/view?usp=drive_link'
path = 'https://drive.google.com/uc?export=download&id='+link.split('/')[-2]
df = pd.read_csv(path)
```

```python
df
```

```python
for i in range(1, 1001):
    df[f'depth_{i}'] = df.filter(like=f'depth-{i}_iter').sum(axis=1)
```

```python
for i in range(1, 1001):
    df.drop(columns=df.filter(like=f'depth-{i}_iter').columns, inplace=True)
```

```python
df.describe()
```

```python
depth_columns = [col for col in df.columns if 'depth_' in col]
```

```python
depth_columns
```

```python
df['Depth sequencing'] = df[depth_columns].sum(axis=1)
```

```python
df['Depth sequencing']
```

```python
df
```

```python
for column in depth_columns:
    median_column = f"{column}_median"
    q1_column = f"{column}_q1"
    q3_column = f"{column}_q3"
    df[median_column] = np.median(df[column])
    df[q1_column] = np.percentile(df[column], 25)
    df[q3_column] = np.percentile(df[column], 75)

```

```python
df
```

```python
object_columns = df.select_dtypes(include=['object']).columns
object_columns = object_columns.drop('sample-id', errors='ignore')
df1 = df.drop(columns=object_columns)
```

```python
df_melted = pd.melt(df1, id_vars=['sample-id'], var_name='depth_stat', value_name='value')

```

```python
df_melted
```

```python
df_melted['depth_number'] = df_melted.loc[df_melted['depth_stat'].str.startswith('depth_'), 'depth_stat'].str.extract(r'(\d+)').astype(int)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style
sns.set(style="whitegrid")

# Create a boxplot
plt.figure(figsize=(12, 8))  # Adjust the figure size
sns.boxplot(x='depth_number', y='value', data=df_melted, showfliers=True)
plt.xlabel('Depth Number', fontsize=14)  # Increase font size
plt.ylabel('Value', fontsize=14)  # Increase font size
plt.title('Boxplot of Value by Depth Number', fontsize=16)  # Increase font size
plt.xticks(rotation=45, fontsize=12)  # Rotate x-axis labels and increase font size
plt.tight_layout()

# Show the plot
plt.show()
```
